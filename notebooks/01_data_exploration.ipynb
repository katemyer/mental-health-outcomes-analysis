```python
#Question: What factors are associated with homicide clearance (solved vs unsolved)?
```

```python
#inspect the target variable -> Solved
df["Solved"].value_counts(dropna=False)
```

```python
#convert to numerical 1 = solved, 0 = unsolved
df["Solved"] = df["Solved"].map({"Yes": 1, "No": 0})
df["Solved"].mean()
```

```python
#solve rate by weapon - which one has the lowest solve rates
df.groupby("Weapon")["Solved"].mean().sort_values().head(10)
```

```python
#solve rate by victim sex
df.groupby("VicSex")["Solved"].mean().sort_values()
```

```python
#solve rate by rate
df.groupby("VicRace")["Solved"].mean().sort_values()
```

```python
#solve rate by state
df.groupby("State")["Solved"].mean().sort_values().head(10)
```

```python
#solve rate trend over time
df.groupby("Year")["Solved"].mean()
```
